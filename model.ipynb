```python
import ast
tokens =[]
def convert(obj):
    for i in ast.literal_eval(obj):
        tokens.append(i['name'])
    return(tokens)
        
```

```python
movies.genres = movies.genres.apply(convert)
```

```python
movies.genres
```

```python
movies.head()
```

```python
movies.keywords = movies.keywords.apply(convert)
```

```python
movies.head()
```

```python
movies.cast[0]
```

```python
# taking out first 3 cast actors from cast column

def convert2(obj):
    cast_list =[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            cast_list.append(i['name'])
            counter+=1
        else:
            break    
    return(cast_list)
        
```

```python
movies.cast = movies.cast.apply(convert2)
```

```python
movies.head()
```

```python
movies.crew[0]
```

```python
# Ftechongg movie director 
def fetch_director(obj):
    dir_list =[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            dir_list.append(i['name'])
            break    
    return(dir_list)
```

```python
movies.crew = movies.crew.apply(fetch_director)
movies.crew
```

```python
movies.head()
```

```python
movies.overview
```

```python
# overview column is string, let us convert it inot list
movies.overview = movies.overview.apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# let us concatenate all the last list columns to make one single column.
# before that let us remove all the spaces between all words eg. Science Fiction - make it ScienceFiction
movies.genres = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.genres
```

```python
movies.keywords = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies.cast = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies.crew = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#now let us combine all these columns into 1 and create a new column with name "Tag"
movies['tags'] = movies['overview']+ movies['genres']+ movies['keywords']+ movies['cast']+movies['crew']
```

```python
movies.tags
```

```python
movies.head()
```

```python
new_df = movies[['id','title','tags']]
```

```python
new_df['tags'][0]
```

```python
# converting tags list into string
new_df['tags'] = new_df['tags'].apply(lambda x:' '.join(x))
```

```python
new_df.head()
```

```python
#convert tag column in lowercase
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.info()
```

```python
new_df.head()
```

```python
# let us do vectorization and convert the tags column text into vectors using BOW by remvoing the stopwords
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
cv.get_feature_names()
```

```python
#In feature names, there are many words which has similar meaning like angel, angels or anger, angry etc.
# so, we are required to perform STEMMING. let us down the library first
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
new_df.tags
```

```python
ps.stem('loving')
```

```python
# let us create a function to perform stemming
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
        
    
```

```python
new_df.tags = new_df.tags.apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
# Now we have evrythign ready, movie titles which are most similar. 
# Let us create recommend function
def recommend(movie):
    #first find index and the related title of it
    movie_index = new_df[new_df['title']==movie].index[0]
    distance = similarity[movie_index]
    #sorting, by keeping index value
    movie_list = sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```
